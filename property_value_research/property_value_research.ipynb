### Заполение пропусков

```python
df.isna().sum() #Подсчет пропущенных значений
```

```python
def nafill(column,func): #создание функции для замены пропущенных значений
    df[column] = df[column].fillna(func)
```

```python
nafill('balcony',0) #замена пропущенных значений в столбце на 0
nafill('parks_around3000',0)
nafill('ponds_around3000',0)
nafill('is_apartment', False) #Замена пропущенных значений на False
df.dropna(subset=['floors_total','locality_name'],inplace=True)#Удаление строк где отсутствуют значения в столбцах locality_name и floors_total
df.isna().sum()
```

Первые пропуски были заполнены, остальные столбцы, нужно будет заменять определенными значениями. Для заполнения пропусков в `living_area` и `kitchen_area` мы можем посчитать сколько в среднем в недвижимостях площади занимают эти части и вставить в пропущеные значения части от общей площади. 
Значения в столбце `ceiling floor` можно заменить на средние или медианные.
Остальные значения придется оставить пропущенными, но они не повлияют на исследование

```python
def room_part(column):
    return df[column].mean()/df['total_area'].mean() #Функция для подсчета средней части комнаты от общей площади
df['living_area'].fillna(df['total_area']*room_part('living_area'), inplace=True) #Заполнение пропусков в площадях комнат в зависимости от
df['kitchen_area'].fillna(df['total_area']*room_part('kitchen_area'),inplace=True) # их средней части от средней общей площади
df.isna().sum()
```

```python
print(df['ceiling_height'].mean())
```

Средняя высота потолков равняется 2.77 метрам

```python
df['ceiling_height'].fillna(df['ceiling_height'].median(), inplace=True) #Замена всех пропущеных значений на медиану
print(df['ceiling_height'].mean())
display(df.head())
```

После замены пропущеных значений в столбце высоты потолков на медианные средняя высота изменилась на 5 см, что не критично в нашем анализе и мы можем оставить эти данные в таком виде

## Расчёты и добавление результатов в таблицу

Для удобного анализа и получения большей информации в таблицу было решено добавить столбцы со значениями цены квадратного метра, дня недели, месяца и года публикации, этажа квартиры (с разделением первый,последний,другой), соотношение жилой и общей площади и соотношение кухни и общей площади

```python
df['cost_per_m'] = df['last_price']/df['total_area'] #создание столбца с расчетом цены квадратного метра недвижимости делением цены на площадь
df['weekday'] = df['first_day_exposition'].dt.weekday #Создание столбцов со значениями дня недели, месяца и года
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
df['living_part'] = df['living_area']/df['total_area'] #Создание столбцов со значениями соотношения площадей частей к общей площади
df['kitchen_part'] = df['kitchen_area']/df['total_area']
def floor_typing(row): #Создание функции для определения типа этажа
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
df['floor_type'] = df.apply(floor_typing, axis=1)
display(df.head())
```

По итогу добавления столбцов: 
* `cost_per_m` - цена квадратного метра недвижимости
* `weekday` - день недели, в который было опубликовано объявление
* `month` - месяц, в который было опубликовано объявление
* `year` - год, в который было опубликовано объявление
* `living_part` - соотношение жилой площади к общей площади
* `kitchen_part` - соотношение площади кухни к жилой площади
* `floor_type` - этаж (первый, последний или другой)

### Определение центра города

```python
df_spb = df.query('locality_name == "Санкт-Петербург"') #Срез датафрейма по городу Санкт-Петербург
df_spb = df_spb[['locality_name','last_price','cityCenters_nearest']] #Сохранение лишь нужных для анализа столбцов
df_spb['cityCenters_km'] = df_spb['cityCenters_nearest']/1000 #Перевод из метров в км и добавление значения в новый столбцец
df_spb = df_spb.dropna() #Удаление строк где отсутствуют значения
df_spb['cityCenters_km'] = df_spb['cityCenters_km'].astype(int) #Перевод столбца с рассточнием в км к типу integer

df_spb['cost_per_km'] = df_spb['last_price']/df_spb['cityCenters_km']
display(df_spb.head())

df_spb = df_spb.pivot_table(index='cityCenters_km',values='cost_per_km')
df_spb.plot(grid=True,figsize=(10,8),xlim=(0,15))
```

По графику видно, что цена начинает расти сильнее обычного на 6 км и чем ближе к центру города, тем дороже становится стоимость. Исходя из этого мы можем сделать вывод, что центральная зона города начинается на 6 км.
